```python
import pandas as pd
df = pd.read_csv('Titanic-Dataset.csv')
df
```

```python
df.head()
```

```python
dfc= df.dropna()
dfc
```

```python
df['Age'].fillna(df['Age'].mean(), inplace= True)
df
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace= True)
df
```

```python
df.isnull().sum()
```

```python
df['Fare_Normalized'] = (df['Fare']- df['Fare'].mean()) / df['Fare'].std()
df
```

```python
df[['Fare','Fare_Normalized']].head()
```

```python
outliers = df[df['Fare'] > 300]
outliers
```

```python
df_no_outliers = df[df['Fare'] <=300]
df_no_outliers
```

```python
df_no_outliers[['Fare']].head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#missing data before cleaning

sns.heatmap(df.isnull(), cbar=True, cmap="viridis")
plt.title("Missing values before cleaning")
plt.show()
```

```python
sns.heatmap(dfc.isnull(), cbar=True, cmap="viridis")
plt.title("Missing values after cleaning")
plt.show()
```

```python
#visualize the fare distributiom before & after removing outliers

plt.figure(figsize=(10,5))
sns.histplot(df['Fare'],bins=50,kde= True)
plt.title('Fare distribution with outliers')
plt.show()

```

```python

plt.figure(figsize=(10,5))
sns.histplot(df_no_outliers['Fare'],bins=50,kde= True)
plt.title('Fare distribution without outliers')
plt.show()
```
