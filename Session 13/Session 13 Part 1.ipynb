```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Titanic-Dataset.csv')
df
```

```python
df.head(3)
```

```python
class_counts = df['Pclass'].value_counts()
class_counts
```

```python
plt.figure(figsize=(6, 4))
plt.bar(class_counts.index, class_counts.values,color='pink')
plt.title('Passenger class distribution ( Matplotlib)')
plt.xlabel('Class')
plt.ylabel('No. of Passengers')
plt.show()
```

**Session 13 A**

```python
#customizing the bar chart using Matplotlib

plt.figure(figsize=(6, 4))
plt.bar(class_counts.index, class_counts.values,color= ['#D8D2C2','#B17457','#4A4947'])
plt.title('Passenger class distribution ( Matplotlib)')
plt.xlabel('Class')
plt.ylabel('No. of Passengers')
plt.xticks([1,2,3],['First Class', 'Second Class', 'Third Class'])
plt.show()
```

**Session 12**

```python
import seaborn as sns

plt.figure(figsize=(8,5))
sns.boxplot(x= "Survived", y ="Age", data = df, palette= 'Set3')
plt.title('Age distribution by survival status using seaborn')
plt.xlabel('Survived ( 1= Yes, 0 = No)')
plt.ylabel('Age')
plt.show()
```

**Session 13 B**

```python
#customizing seaborn box-plot

plt.figure(figsize=(8,5))
sns.boxplot(x= "Survived", y ="Age", data = df, palette= 'flare')
plt.title('Age distribution by survival status using seaborn')
plt.xlabel('Survived ( 1= Yes, 0 = No)')
plt.ylabel('Age')
plt.xticks([0,1],['Did Not Survive', 'Survived'])
plt.show()
```
